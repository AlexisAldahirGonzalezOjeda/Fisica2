# 1 Gráficas de posición, velocidad y aceleración en un MAS.

Grafica de una particula que se mueve con Movimiento Armonico Simple.

$x = A∗sin(\omega∗t + \alpha)$ 

Teniendo:

Amplitud (A) = 0.3metros

Frecuencia angular ($\omega$) = $\pi$

Fase inicial ($\alpha$) =  0

# 1.1 Gráfica de posición

$x = 0.3∗sin(\pi∗t)$ 

```python
import numpy as np 
import matplotlib.pyplot as plt

x=np.arange(0,10,0.1)
y = 0.3*np.sin(3.1415*x)
plt.plot(x,y)
plt.xlabel("t(s)", fontsize = 20)
plt.ylabel("x(m)", fontsize = 20)
plt.grid(True)
plt.show()
```

# 1.2 Gráfica de velocidad

$v = \pi*0.3∗cos(\pi∗t)$ 

```python
import numpy as np 
import matplotlib.pyplot as plt

x=np.arange(0,10,0.1)
y = 3.1415*0.3*np.cos(3.1415*x)
plt.plot(x,y)
plt.xlabel("t(s)", fontsize = 20)
plt.ylabel("v(m/s)", fontsize = 20)
plt.grid(True)
plt.show()
```

# 1.3 Gráfica de aceleración

$a = -\pi^2*0.3∗sin(\pi∗t)$ 

```python
import numpy as np 
import matplotlib.pyplot as plt

x=np.arange(0,10,0.1)
y = -3.1415*3.1415*0.3*np.sin(3.1415*x)
plt.plot(x,y)
plt.xlabel("t(s)", fontsize = 20)
plt.ylabel("a(m/s*s)", fontsize = 20)
plt.grid(True)
plt.show()
```

# 1.4 Análisis de las Gráficas de posición, velocidad y aceleración en un MAS

```python
import numpy as np 
import matplotlib.pyplot as plt

x=np.arange(0,10,0.1)
y1= 0.3*np.sin(3.1415*x)
plt.plot(x,y1)
y2= 3.1415*0.3*np.cos(3.1415*x)
y3= -3.1415*3.1415*0.3*np.sin(3.1415*x)
plt.plot(x,y2)
plt.plot(x,y3)
plt.legend(('x', 'v', 'a'), loc = 'upper right')
plt.xlabel("t(s)", fontsize = 20)
plt.ylabel("x(m)", fontsize = 20)
plt.grid(True)
plt.show() 
```

Al sobreponer las gráficas de posición, velocidad y aceleración en un MAS, podemos observar que la aceleracion (en verde) es proporcional y opuesta al desplazamiento (en azul).

# 2 Gráficas de los diferentes tipos de movimiento oscilatorio amortiguado

# 2.1 Oscilaciones sobreamortiguadas

```python
import numpy as np 
import matplotlib.pyplot as plt

x=np.arange(0,10,0.1)
y1= np.exp(-0.5*x)
plt.plot(x,y1)
y2= np.exp(-2*x)
y3= np.sin(2*x)
plt.plot(x,y3)
plt.plot(x,y2)
y = np.exp(-0.5*x) + np.exp(-2*x)
plt.plot(x,y)
plt.legend(('exp(-0.5*x)', 'exp(-2*x)', 'exp(-0.5*x) + exp(-2*x)'), loc = 'upper right')
plt.xlabel("t(s)", fontsize = 20)
plt.ylabel("x(m)", fontsize = 20)
plt.grid(True)
plt.show()
```

# 2.2 Oscilaciones subamortiguadas

```python
import numpy as np 
import matplotlib.pyplot as plt

x=np.arange(0,0.7,0.01)
y1 = 5.01*np.exp(-7*x)*np.sin(99.75*x+1.5)
plt.plot(x,y1)
y2 = 5.01*np.exp(-7*x)
plt.plot(x,y2)

plt.legend(('a'), loc = 'upper right')
plt.xlabel("t(s)", fontsize = 20)
plt.ylabel("x(m)", fontsize = 20)
plt.grid(True)
plt.show()
```

# 3 Gráficas de energía
