# 1 Gráficas de posición, velocidad y aceleración en un MAS.

Grafica de una particula que se mueve con Movimiento Armonico Simple.

$x = A∗sin(\omega∗t + \alpha)$ 

Teniendo:

Amplitud (A) = 0.3metros

Frecuencia angular ($\omega$) = $\pi$

Fase inicial ($\alpha$) =  0

# 1.1 Gráfica de posición

$x = 0.3∗sin(\pi∗t)$ 

```python
import numpy as np 
import matplotlib.pyplot as plt

x=np.arange(0,10,0.1)
y = 0.3*np.sin(3.1415*x)
plt.plot(x,y)
plt.xlabel("t (s)", fontsize = 20)
plt.ylabel("x (m)", fontsize = 20)
plt.grid(True)
plt.show(True)
```

# 1.2 Gráfica de velocidad

$v = \pi*0.3∗cos(\pi∗t)$ 

```python
import numpy as np 
import matplotlib.pyplot as plt

x=np.arange(0,10,0.1)
y = 3.1415*0.3*np.cos(3.1415*x)
plt.plot(x,y)
plt.xlabel("t (s)", fontsize = 20)
plt.ylabel("v (m/s)", fontsize = 20)
plt.grid(True)
plt.show()
```

# 1.3 Gráfica de aceleración

$a = -\pi^2*0.3∗sin(\pi∗t)$ 

```python
import numpy as np 
import matplotlib.pyplot as plt

x=np.arange(0,10,0.1)
y = -3.1415*3.1415*0.3*np.sin(3.1415*x)
plt.plot(x,y)
plt.xlabel("t (s)", fontsize = 20)
plt.ylabel("a (m/s^2)", fontsize = 20)
plt.grid(True)
plt.show()
```

# 1.4 Análisis de las Gráficas de posición, velocidad y aceleración en un MAS

```python
import numpy as np 
import matplotlib.pyplot as plt

x=np.arange(0,10,0.1)
y1= 0.3*np.sin(3.1415*x)
plt.plot(x,y1)
y2= 3.1415*0.3*np.cos(3.1415*x)
y3= -3.1415*3.1415*0.3*np.sin(3.1415*x)
plt.plot(x,y2)
plt.plot(x,y3)
plt.legend(('x', 'v', 'a'), loc = 'upper right')
plt.xlabel("t (s)", fontsize = 20)
plt.ylabel("x (m)", fontsize = 20)
plt.grid(True)
plt.show() 
```

Al sobreponer las gráficas de posición, velocidad y aceleración en un MAS, podemos observar que la aceleracion (en verde) es proporcional y opuesta al desplazamiento (en azul).

# 2 Gráficas de los diferentes tipos de movimiento oscilatorio amortiguado

Las características esenciales de las oscilaciones amortiguadas:

• La amplitud de la oscilación disminuye exponencialmente con el tiempo.

• La energía del oscilador también disminuye, debido al trabajo de la fuerza Fr de rozamiento viscoso opuesta a la velocidad.


# 2.1 Oscilaciones sobreamortiguadas

se produce cuanto hay un amortiguamiento grande, en este caso no hay oscilaciones, la partiula se aproxima a la posicion de equilibrio.

```python
import numpy as np 
import matplotlib.pyplot as plt

x=np.arange(0,1,0.005)
y1 = 10*np.exp(-15*x)*np.sin(8*x+1.5)
plt.plot(x,y1)
plt.legend(('exp(-0.5*x)'), loc = 'upper right')
plt.xlabel("t (s)", fontsize = 20)
plt.ylabel("x (m)", fontsize = 20)
plt.grid(True)
plt.show()
```

# 2.2 Oscilaciones subamortiguadas

```python
import numpy as np 
import matplotlib.pyplot as plt

x=np.arange(0,0.7,0.005)
y1 = 5.01*np.exp(-7*x)*np.sin(99.75*x+1.5)
plt.plot(x,y1)
y2 = 5.01*np.exp(-7*x)
plt.plot(x,y2)

plt.legend(('a','5.01e^-7x'), loc = 'upper right')
plt.xlabel("t (s)", fontsize = 20)
plt.ylabel("x (m)", fontsize = 20)
plt.grid(True)
plt.show()
```

# 4 Superposición de M.A.S

## 4.1 Igual direccion, igual frecuencia

Al sumar los movimientos el resultado es un M.A.S. de la misma dirección y
de la misma frecuencia.

Si los dos M.A.S. están en fase, la diferencia de fase es cero, la amplitud del M.A.S
resultante es la suma de las amplitudes de los dos M.A.S.

Si los dos M.A.S. están en oposición de fase, la diferencia de fase es 180, la amplitud
del M.A.S resultante es la diferencia de las amplitudes de los dos M.A.S. 

```python
import numpy as np 
import matplotlib.pyplot as plt

w=4
A1=5
A2=7
f1=0
f2=3.14159
d=f2-f1
A=np.sqrt(A1*A1+A2*A2+2*A1*A2*np.cos(d))

t=np.arange(0,3,0.005)
x1 = A1*np.sin(w*t+f1)
plt.plot(t,x1)

x2 = A2*np.sin(w*t+f2)
plt.plot(t,x2)

X = A*np.sin(w*t+d)
plt.plot(t,X)

plt.legend(('x1','x2','x1+x2'), loc = 'upper left')
plt.xlabel("t (s)", fontsize = 20)
plt.ylabel("x (m)", fontsize = 20)
plt.grid(True)
plt.show()
```

## 4.2 Igual direccion, diferente frecuencia

al sumar dos MAS de misma direccion pero difererente frecuencia angular e igual fase inicial se obtiene un movimiento el cual no cumple con las caracteristicas de un MAS. 


```python
import numpy as np 
import matplotlib.pyplot as plt

A=0.2
w1=19.0
w2=15.0

t=np.arange(0,10,0.005)

X = 2*A*np.cos((w1-w2)/2*t)*np.sin((w1+w2)/2*t)
plt.plot(t,X)

x = 2*A*np.cos((w1-w2)/2*t+6.2832)
plt.plot(t,x)

plt.legend(('X','amplitud modulada'), loc = 'upper left')
plt.xlabel("t (s)", fontsize = 20)
plt.ylabel("x (m)", fontsize = 20)
plt.grid(True)
plt.show()
```

### 4.3 Direcciones perpendiculares, igual frecuencia

Al sumar dos MAS con direcciones perpendiculares se obtiene las figuras de Lissajous, las cuales dependen de la diferencia de fase y la relacion w1/W2

```python
from numpy import sin,pi,linspace
from pylab import plot,show,subplot

delta = [0,pi/2,3*pi/2,pi]
t = linspace(-pi,pi,300)

for i in range(0,4):
 x = 4*sin(2* t + delta[i])
 y = 3*sin(2 * t)
 subplot(2,2,i+1)
 plot(x,y)
 
show()
```
